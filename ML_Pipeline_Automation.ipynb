```python
import sqlite3
import pandas as pd

conn = sqlite3.connect("test.db")

# Create sample table
df = pd.DataFrame({
    "id": [1, 2, 3],
    "name": ["Alice", "Bob", "Charlie"],
    "sales": [1200, 900, 1450]
})

df.to_sql("sales_table", conn, if_exists="replace", index=False)

conn.close()
```

```python
import sqlite3
import pandas as pd

def run_sql_query(db_path, query):
    conn = sqlite3.connect(db_path)
    result = pd.read_sql_query(query, conn)
    conn.close()
    return result

query = "SELECT * FROM sales_table;"
output = run_sql_query("test.db", query)
output
```
